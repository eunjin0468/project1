```python
# Step 1: 파일 생성
import time
import sys
import os

f_dir = 'I:\\web_crawling\\youtube_files\\'

n = time.localtime()

s = '%04d-%02d-%02d-%02d-%02d-%02d' %(n.tm_year, n.tm_mon, n.tm_mday, n.tm_hour, n.tm_min, n.tm_sec)

os.makedirs(f_dir+s)
os.chdir(f_dir+s) 

fc_name=f_dir+s+'\\'+s+'.csv'
```

```python
# Step 2: 크롬 시크릿모드 창 띄우기
from bs4 import BeautifulSoup
from selenium import webdriver

chrome_options = webdriver.ChromeOptions()  # webdriver의 크롬 옵션 객체 생성
chrome_options.add_argument("--incognito")  # 크롬 옵션에 시크릿 모드 추가
# 위에서 만든 크롬 옵션을 적용하여 크롬창 생성
driver = webdriver.Chrome("I:\\web_crawling\\chromdriver\\chromedriver.exe", options=chrome_options) 

driver.get("https://www.youtube.com/")
time.sleep(2)
```

```python
# Step 5: top10 키워드를 검색하여 나온 영상의 댓글 크롤링 후 각각의 파일로 저장
#words_1 = {'부모님'}
for i in words_10 :
    query_txt = i
    driver.get("https://www.youtube.com/results?search_query="+query_txt)
    
    soup = BeautifulSoup(html, 'html.parser')
    for i in range(1,3) : # defalut 정렬인 '관련성' 기준으로 나열된 영상 중 1,2번째 영상 사용
        streaming = None
        driver.implicitly_wait(10)
        print(i)
        xpath = driver.find_element_by_xpath("""/html/body/ytd-app/div/ytd-page-manager/ytd-search/div[1]/ytd-two-column-search-results-renderer/div/ytd-section-list-renderer/div[2]/ytd-item-section-renderer/div[3]/ytd-video-renderer[%s]/div[1]/div/div[1]/div/h3/a"""%i).click()
        driver.implicitly_wait(4)
        time.sleep(3)                                      
        streaming = soup.find('yt-formatted-string', class_='style-scope ytd-toggle-button-renderer')
        print('streaming: ', streaming) 
        # 실시간 streaming 영상의 태그 element는 이외에도 다양하므로 크게 의미있는 코드는 아님.
        if str(streaming) == '<yt-formatted-string class="style-scope ytd-toggle-button-renderer" id="text">채팅 숨기기</yt-formatted-string>':
            time.sleep(5)
            driver.back()
            #print('back')
            driver.implicitly_wait(5)
            continue

        else:
            driver.implicitly_wait(5)
            last_page_height = driver.execute_script("return document.documentElement.scrollHeight")
            # 처음에 조금만 내려서 댓글창이 로딩될 수 있도록 함. 
            driver.execute_script("window.scrollTo(0, 250);")
            
            while True:
                time.sleep(3)
                # 끝까지 스크롤 다운
                driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")
                driver.implicitly_wait(5)
                new_page_height = driver.execute_script("return document.documentElement.scrollHeight")
                if new_page_height == last_page_height:
                    break
                last_page_height = new_page_height

            time.sleep(3)
            html_source = driver.page_source
            soup = BeautifulSoup(html_source, 'html.parser')

            comment_list = soup.select("yt-formatted-string#content-text")
            #print('comment_list: ', comment_list)

            comment_final = []
            fc_name2 = []
            # 총 10개의 파일 생성(스트리밍 영상만 검색된 loop는 댓글이 없으므로 
            # 저장되지 않아 10개 이하일 수도 있음)
            for i in range(len(comment_list)):
                fc_name2=f_dir+s+'\\'+query_txt+'_검색후댓글.csv'
                f = open(fc_name2, 'a', encoding="UTF-8-sig") 

                temp_comment = comment_list[i].text
                temp_comment = temp_comment.replace('\n', '')
                temp_comment = temp_comment.replace('\t', '')
                temp_comment = temp_comment.replace('    ', '')
                comment_final.append(temp_comment)
                
                f.write(temp_comment)
                f.write("\n")
                f.close()
            
        driver.back() 
```

```python
# Step 6: top10 영상 댓글 크롤링이 저장된 파일들 내용을 가져와서
#         빈도수를 기준으로 각 파일들의 키워드 추출 -> 새로운 키워드 제시
import pandas as pd
from krwordrank.word import KRWordRank
from krwordrank.hangle import normalize

new_keywords = []

# 검색 후 댓글.csv 파일명 리스트 가져오기
path = f_dir+s+'\\'
file_list = os.listdir(path)
file_list_py = [file for file in file_list] #if file.endswith('.csv')]
#print('file_list_py:',file_list_py)

# csv 파일들을 DataFrame으로 불러와서 새로운 키워드 생성
df = pd.DataFrame()
for i in file_list_py[1:]:
    print('\ni:',i)
    fc_name3=path+i
    
    # 키워드 추출
    stopwords = {'수있어서', '제가', '우리가', '있는', '마주한', '반응은', '너무', '진짜', 'the'}
    
    texts = open(fc_name3, encoding='UTF-8').readlines( )
    norm = [normalize(text, english=True, number=False) for text in texts]

    keywords = summarize_with_keywords(norm, min_count=5, max_length=10,
        beta=0.85, max_iter=10, stopwords=stopwords, verbose=True)

    for word, r in sorted(keywords.items(), key=lambda x:x[1], reverse=True)[:1]:
        print('%8s:\t%.4f' % (word, r))
        new_keywords.append(word)
        
print(new_keywords) # 새로운 키워드 생성  
        
df = df.reset_index(drop = True)
```
